# LuluCare 360 — Module 3 | `shared/enums.py`
### Single Source of Truth for All Enum Constants

**Project:** LuluCare 360 — AI Complaint-Resolution Co-Pilot  
**Module:** Shared Enums (used by ALL 4 modules)  
**File maps to:** `shared/enums.py`  

---

> **Why this file exists:**  
> Four people build four modules independently. If each re-types a literal like `"Likely_Abuser"` or `"Acknowledge"`, a single casing slip silently breaks branch matching at integration time *(Trap 17 in the implementation plan)*. Every module imports these constants instead of hard-coding strings.

**Casing is FIXED — do not change:**
| Category | Casing |
|---|---|
| `genuineness`, `claim_status`, `action`, `refund_type` | `UPPER_SNAKE` |
| `issue_type`, `loyalty_tier` | `TitleCase` |
| `frustration` | `Low` \| `Medium` \| `High` |

## 1. Genuineness Constants
*(Investigator output → feeds Economist)*

```python
# --- Genuineness (Investigator output) -------------------------------------------
# These are the three trust verdicts. UPPER_SNAKE casing is fixed.
GENUINE       = "GENUINE"
SUSPICIOUS    = "SUSPICIOUS"
LIKELY_ABUSER = "LIKELY_ABUSER"

GENUINENESS = (GENUINE, SUSPICIOUS, LIKELY_ABUSER)

print("Genuineness constants:", GENUINENESS)
```

## 2. Claim Status Constants
*(Investigator output → feeds Economist)*

```python
# --- Claim status (Investigator output) ------------------------------------------
# CONFIRMED  : agent note backs the claim  → honour exactly what was logged
# CONTRADICTED: agent note contradicts it  → respond from record, deny claimed remedy
# UNVERIFIED : no record of claimed contact → proceed on verified facts only
CONFIRMED    = "CONFIRMED"
CONTRADICTED = "CONTRADICTED"
UNVERIFIED   = "UNVERIFIED"

CLAIM_STATUS = (CONFIRMED, CONTRADICTED, UNVERIFIED)

print("Claim status constants:", CLAIM_STATUS)
```

## 3. Action Constants
*(Economist output → feeds Voice)*

```python
# --- Action (Economist output) ---------------------------------------------------
# The 5 possible resolution actions the Economist can choose.
ACTION_ACKNOWLEDGE   = "ACKNOWLEDGE"    # No payout, polite reply
ACTION_COUPON        = "COUPON"         # Percentage discount coupon
ACTION_WALLET_CREDIT = "WALLET_CREDIT"  # Flat AED wallet top-up
ACTION_REFUND        = "REFUND"         # Full refund
ACTION_ESCALATE      = "ESCALATE"       # Hand to human agent

ACTIONS = (
    ACTION_ACKNOWLEDGE,
    ACTION_COUPON,
    ACTION_WALLET_CREDIT,
    ACTION_REFUND,
    ACTION_ESCALATE,
)

# Actions that send the customer an email. ACKNOWLEDGE and ESCALATE NEVER email.
EMAIL_ACTIONS = (ACTION_COUPON, ACTION_REFUND, ACTION_WALLET_CREDIT)

print("All actions    :", ACTIONS)
print("Email-firing   :", EMAIL_ACTIONS)
print("No-email actions: ACKNOWLEDGE, ESCALATE")
```

## 4. Refund Logistics Constants
*(Economist output → feeds Voice)*

```python
# --- Refund logistics (Economist output) ----------------------------------------
# PICKUP    : send a courier to collect the item (resalable goods)
# KEEP_ITEM : let the customer keep/dispose of it (perishable / not worth freight)
# NONE      : no refund was issued, logistics irrelevant
REFUND_PICKUP    = "PICKUP"
REFUND_KEEP_ITEM = "KEEP_ITEM"
REFUND_NONE      = "NONE"

REFUND_TYPES = (REFUND_PICKUP, REFUND_KEEP_ITEM, REFUND_NONE)

print("Refund types:", REFUND_TYPES)
```

## 5. Frustration Constants
*(Reader output → feeds Economist)*

```python
# --- Frustration (Reader output) ------------------------------------------------
# NOTE: Mixed case (NOT upper-snake). Exactly as stored in messages.csv.
FRUST_LOW    = "Low"
FRUST_MEDIUM = "Medium"
FRUST_HIGH   = "High"

FRUSTRATION = (FRUST_LOW, FRUST_MEDIUM, FRUST_HIGH)

print("Frustration levels:", FRUSTRATION)
```

## 6. Loyalty Tier Constants
